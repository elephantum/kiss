```python
f = file('8728.json')
```

```python
import re
fix_json = re.compile(r'\\(\d+)')
```

```python
import json

keys = set()
for line_raw in f:
    line = fix_json.sub(r'x\1', line_raw)
    d = json.loads(line)
    keys = keys.union(set(d.keys()))
```

```python
columns_def = ',\n    '.join(['`{0}` string'.format(i.replace(' ', '_')) for i in sorted(keys)])
columns = ',\n'.join(['`{0}`'.format(i.replace(' ', '_')) for i in sorted(keys)])

templ = '''create table kiss_{{data_source}} (
    dt string,
    {columns_def}
)
stored as rcfile
location 's3://enter-kiss-test/enter_proto_{{data_source}}/kiss_{{data_source}}/';'''

print templ.format(**globals())
```

```python
json_columns = ',\n    '.join(['"{0}"'.format(i) for i in sorted(keys)])
parsed_columns = ',\n    '.join(['`{0}`'.format(i.replace(' ', '_')) for i in sorted(keys)])
parsed_columns_with_prefix = ',\n    '.join(['`parsed.{0}`'.format(i.replace(' ', '_')) for i in sorted(keys)])

templ = '''insert overwrite table kiss_{{data_source}}
select
    from_unixtime(cast(parsed.t as int)) as dt, 
    {parsed_columns_with_prefix}
from kiss_raw 
lateral view json_tuple(regexp_replace(kiss_raw.json_data, '\\\\', 'x'), 
    {json_columns})
parsed as
    {parsed_columns};'''

print templ.format(**globals())
```
